### Project development capacity requirements

```python
%matplotlib inline
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
D=pd.read_excel("/Users/annachystiakova/Documents/Employee forecast/Loe_projects_required.xlsx", sheetname='loe_plan')
D.columns=["Q",	"Role",	"Summary required",	"Summary commited",	"RTB%",	"Overhead%",	"Projects"]
D.groupby(['Q','Role'])[['Summary required', 'Summary commited']].sum().plot(kind='bar')
plt.title("Number of staff for project activity FY15")
D['LOE_diff']=D['Summary required']- D['Summary commited']
D['Required_Staff_for_projects']=np.int32(((D['LOE_diff']*100/D['Projects'])/13).tolist())
print("*Summary required and *Summary commited values are provided in weeks for Q")
```

```python
fig, axes = plt.subplots(4, 2, figsize=(30, 30))
roles=pd.unique(D.Role.values)
axes = axes.ravel()
for x, city in enumerate(D.Role.value_counts().index.values):
    data = D[(D['Role'] == city)]
    data.index=[str(data.ix[x].Q)+"-"+str(data.ix[x].Role) for x in data.index]
    data=data.Required_Staff_for_projects
    left=  [k[0] for k in enumerate(data)]
    right=  [k[1] for k in enumerate(data)]

    axes[x].bar(left,right,label="%s" % (city))
    axes[x].set_xticks(left, minor=False)
    labels=[]
    for i in data.index.values:
        if i in labels:
            labels.append("")
        else:
            labels.append(i)
    axes[x].set_xticklabels(labels, rotation=10)
    axes[x].set_ylabel('Required staff')
   
    axes[x].legend(loc='best')
    axes[x].grid(True)
fig.suptitle('Number of required staff to complete all projects for FY15', fontsize=30)

#D[D.Role=='Admins'].LOE_diff.plot(kind='bar')
```
